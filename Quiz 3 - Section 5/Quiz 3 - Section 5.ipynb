the modified "positionText" column from string to numeric <br>
using ```pd.to_numeric()```

HINT: See the example in Lecture 13

````python
# Write your own code
df = pd.read_csv('data_raw/sprint_results.csv')


unique_values = df['positionText'].unique()
replace_dict = {
    'R': 'np.nan',
    'N': 'np.nan'
}

for value in unique_values:
    df['positionNumeric'] = df['positionText'].replace(value, replace_dict.get(value))


unique_values = df['positionNumeric'].unique()


display(unique_values)


````

(c) Aggregate and query

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Subset the observations for 'constructorId == 3' using ```.query()```
- Obtain the sum of "laps" by "driverId" <br> 
 using ``` .groupby().agg() ```
- Subset the drivers with a sum of laps $\ge 40$
- Store the results of all these operations in a dataframe <br>
```sprint_results_agg ```

````python
df = pd.read_csv('data_raw/sprint_results.csv')


constructorId = df.query('constructorId == 3')

grouped_constructorId = constructorId.groupby('driverId').agg({'laps': 'sum'})

sprint_results_agg = grouped_constructorId.groupby('sum >= 40')

display(sprint_results_agg)

````

(d) Aggregate and sort

<font size = "3">

- Import the "data_raw/sprint_results.csv" file
- Obtain the sum of "laps" by "driverId" <br> 
using ``` .groupby().agg() ```
- Sort the teams in descending order using of <br>
the sum of laps using ``` .sort_values() ```

````python
# Write your own code

df = pd.read_csv('data_raw/sprint_results.csv')

grouped2_df = df.groupby('driverId').agg({'laps': 'sum'})

sorted_df = grouped2_df.sort_values(by='laps', ascending=False)

display(sorted_df)
````
